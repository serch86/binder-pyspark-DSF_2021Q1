```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
M = 50
dim = 2
N = 100
```

```python
for dim in range(2, 101):
    pop = np.random.rand(dim*N).reshape(-1, dim)
    vals = []
    for _ in range(M):
        new_point = np.random.rand(dim).reshape(-1, dim)
        norms = np.linalg.norm(pop-new_point, axis=1, keepdims=True)
        coef = norms.min()/norms.mean()
        vals.append(coef)
    if dim == 2:
        df = pd.DataFrame(vals, columns=[f'N{dim}'])
    else:
        df[f'N{dim}'] = vals
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
df.plot.box(ax=ax)
```
